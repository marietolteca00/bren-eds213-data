# Using a Cursor
For the cursor think about this as a tunnel
We are creating a cursor, we are selecting rows from Site column (a query). Only returning only a few. 
When we fetchall() we are asking the cursor to give us all the data

```python
# Create a cursor
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM Site")
```

```python
# You can apply the fetchall()..

cur.fetchall()
```

Need to recreate the cursor execute line

```python
# This line needs to be re-ran
cur.execute("SELECT * FROM Site")

# Getting one row
cur.fetchone()
```

```python
cur.fetchone()[0]
```

```python
# We want to iterate through the rows and print the site code
cur.execute("SELECT * FROM Site")

for row in cur.fetchall():
    print(row[0])
```

```python
# creating cursor
cur.execute("SELECT * FROM Site")

while True:
    row = cur.fetchone()
    if row == None:
        break
    print(f"Site code: {row[0]}")
```

```python
# You can create a table
cur.execute("""CREATE TEMP TABLE temp_nests AS
            SELECT * FROM Bird_nests""")
```
